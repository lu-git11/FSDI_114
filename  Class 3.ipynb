# link list
1. singly linked list - uni-directional
2. doubly linked list - bi-directional

unlike stack and queue, linked list allow for addition or removal of nodes from anywhere in collection. this means one can replace the "head", insert in the middle of the list, as well as "append" to the list (replace the 'tail').

```python
# from scratch singly linked list

class SinglyLinkedList:
    class __Node:
        def __init__(self, datum):
            self.data = datum
            self.next = None

    def __init__(self):
        self.head = None
        self.tail = None
        self.count = 0
    
    def append(self, value):
        new_node = self.__Node(value)
        self.count += 1
        if not self.tail:
            self.head = new_node
            self.tail = new_node
        else:
            self.tail.next = new_node
            self.tail = new_node

    def insert(self, index, value):
        new_node = self.__Node(value)
        current = self.head
        
        self.count += 1
        if not self.head:
            self.head = new_node
            self.tail = new_node
        elif index == 0:
            new_node.next = self.head
            self.head = new_node
        else:
            current = self.head
            prev = None 
            found = False
            counter = 0 
            while current and not found:
                if counter == index:
                    found = True
                else:
                    counter += 1
                    prev = current
                    current = current.next
                if found:
                    if prev:
                        new_node.next = current
                        prev.next = new_node
                else:
                    prev.next = new_node
                    self.next = new_node
                                    
                        
    def remove(self, value):
        current = self.head
        prev = None
        found = False
        while current:
            if current and not found:
                if current.data == value:
                    found = True
                else:
                    prev = current
                    current = current.next
            if found:
                count -= 1
                if prev:
                    prev.next = current.next
                    if not prev.next:
                        self.tail = prev
                else:
                    self.head = self.head.next
                    if not self.head:
                        self.tail = None 
            else:
                raise ValueError("Value not found")
                        
    def index(self, value):
        counter = 0 
        found = False
        if self.head:
            current = self.head
            while current and not found:
                if current.data == value:
                    found = True
                else:
                    counter += 1
                    current = current.next
        if found:
            return counter
        else:
            raise ValueError("Value not found")

    def __str__(self):
        out = "["
        current = self.head
        if current:
            out += "%s" % current.data
            current = current.next
            while current:
                out += ", %s" % current.data
                current = current.next
        out += "]"
        return out

    def __len__(self):
        return self.count
```
